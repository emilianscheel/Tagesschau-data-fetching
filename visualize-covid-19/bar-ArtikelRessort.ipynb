```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json

df = pd.read_json('../data/database-10-12.json')
```

```python
ressortdf = df.ressort.value_counts().reset_index(name="amount")

ax = ressortdf.plot(kind="bar", x="index", y="amount", figsize=(7,5), legend=False, rot=30, color="#f2cc8f", edgecolor='#e07a5f', title="Anzahl Artikel nach Ressort")
ax.axes.get_xaxis().get_label().set_visible(False)

ax.get_figure().savefig('bar-ressorts.png', bbox_inches="tight")
```
