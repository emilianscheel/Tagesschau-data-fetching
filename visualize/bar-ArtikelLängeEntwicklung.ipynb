```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json('../data/database-27-10.json')
```

```python
df['content_length'] = df['content'].str.len()
df['content_words'] = df['content'].str.split().map(len)
```

```python


#df.loc[(df['date'].dt.year == 2022) & (df['date'].dt.month == 9)].count()
```

```python
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['week'] = df['date'].dt.strftime('%W')
```

```python

dfMonth = df.groupby(['month', 'year']).agg({"content_length": "mean", "content_words": "mean"}).reset_index()
dfWeek = df.groupby(['week']).agg({"content_length": "mean", "content_words": "mean"}).reset_index()
```

```python
dfMonth.plot(x='month', y=['content_words'])
```

```python
# durchschnittslÃ¤nge der Artikel pro Monat
dfWeek.plot(x='week', y=['content_words'])
```
