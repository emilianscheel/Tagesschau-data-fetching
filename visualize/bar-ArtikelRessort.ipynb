```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json('../data/database.json')
```

```python
df['color'] = '#FF0000'

df['ressort'] = df['ressort'].str.capitalize()
```

```python
_df = df[(df['ressort'] != '')]

_df = (_df['ressort']
        .value_counts()
        .rename_axis('ressort')
        .reset_index(name='amount'))

_df['amount_formatted'] = str(_df['amount']) + ' Stück'
        
ax = _df.plot(kind='barh',x='ressort', y='amount', label='absolute Anzahl der Artikel', figsize=(20,10), colormap='Paired')

ax.set_title("Anzahl der Artikel nach Ressort")
ax.set_yticklabels(_df['ressort'])
ax.set_xlabel("absolute Anzahl der Artikel")
ax.set_ylabel("Ressort")
```
