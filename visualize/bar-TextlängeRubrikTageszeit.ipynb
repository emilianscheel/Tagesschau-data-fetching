```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json('../data/database.json')
```

```python
df['hour'] = df['date'].dt.hour

df1 = df.loc[df['ressort'] == 'sport']
df2 = df.loc[df['ressort'] != 'sport']
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, constrained_layout=True, figsize=(10, 5))

df1 = df1.groupby(['hour']).size().reset_index(name='amount')
df2 = df2.groupby(['hour']).size().reset_index(name='amount')

df1.plot(kind="bar", ax=ax1, x='hour', y='amount')
df2.plot(kind="bar", ax=ax2, x='hour', y='amount')

ax1.set_title('Sportartikel nach Veröffentlichungsstunde')
ax2.set_title('Alle außer Sportartikel nach Veröffentlichungsstunde')

ax1.legend(["Anzahl Artikel"])
ax2.legend(["Anzahl Artikel"])

ax1.set_xlabel('Stunde')
ax2.set_xlabel('Stunde')

plt.show()
```
