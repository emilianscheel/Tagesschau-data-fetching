```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json('../data/database.json')
```

```python
df['content_length'] = df['content'].str.len()
df['content_words'] = df['content'].str.split().map(len)
```

# Textl√§nge nach Anzahl Zeichen

```python
fig, ax1 = plt.subplots(1)

plt.rcParams["figure.figsize"] = (30, 10)

df.groupby(['category']).boxplot(column=['content_length'], ax=ax1, subplots=False, rot=45, fontsize=12, showfliers=False)

ax1.set_title('Textl√§nge gemessen an Anzahl Zeichen nach Kategorie')
ax1.set_xticklabels(df['category'].unique(),ha='right')

plt.show()
```

# Textl√§nge nach Anzahl W√∂rter

```python
fig, ax1 = plt.subplots(1)

plt.rcParams["figure.figsize"] = (30, 10)

df.groupby(['category']).boxplot(column=['content_words'], ax=ax1, subplots=False, rot=45, fontsize=12, showfliers=False)

ax1.set_title('Textl√§nge gemessen an Anzahl W√∂rtern nach Kategorie')
ax1.set_xticklabels(df['category'].unique(),ha='right')

plt.show()
```
