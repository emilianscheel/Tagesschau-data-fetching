```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json('../data/database.json')
```

```python
df['content_length'] = df['content'].str.len()
df['content_words'] = df['content'].str.split().map(len)
```

# Textlänge nach Anzahl Zeichen

```python
fig, ax1 = plt.subplots(1)

df.groupby(['ressort']).boxplot(column=['content_length'], ax=ax1, subplots=False, rot=45, fontsize=12, showfliers=False)

ax1.set_xticklabels(['Kein Ressort', 'Ausland', 'Inland', 'Investigativ', 'Sport', 'Wirtschaft'],ha='right')

ax1.set_title('Textlänge gemessen an Anzahl Zeichen nach Ressort')

plt.show()
```

```python
fig, ax1 = plt.subplots()

df.groupby(['ressort']).mean().plot(ax=ax1, kind="bar", y=['content_length', 'content_words'])

ax1.set_title('Durchschnittliche Textlänge nach Ressort')
ax1.legend(["Anzahl Zeichen", "Anzahl Wörter"])
ax1.xaxis.label.set_visible(False)

plt.show()
```

# Textlänge nach Anzahl Wörter

```python
fig, ax1 = plt.subplots(1)

df.groupby(['ressort']).boxplot(column=['content_words'], ax=ax1, subplots=False, rot=45, fontsize=12, showfliers=False)

ax1.set_xticklabels(['Kein Ressort', 'Ausland', 'Inland', 'Investigativ', 'Sport', 'Wirtschaft'])

ax1.set_title('Textlänge gemessen an Anzahl Wörter nach Ressort')

plt.show()
```
