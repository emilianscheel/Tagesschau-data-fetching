```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

news = pd.read_json('../data/database.json')
```

```python
import calendar
import datetime

# Erstellt eine neue Zeile mit dem Wochentag
news['day_of_week'] = news['date'].dt.day_of_week
news['day_of_week_word'] = news['date'].dt.day_name(locale='de_de') 

news['hour'] = news['date'].dt.hour
news['content_length'] = news['content'].str.len()
```

```python
news = news.groupby(["hour"]).agg(amount=('hour', 'size'), content_length_mean=('content_length', 'mean'))
```

```python
# libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Control figure size for this notebook:
plt.rcParams['figure.figsize'] = [20, 10]


 
# use the scatterplot function to build the bubble map
sns.scatterplot(data=news, x="hour", y="day_of_week", size="amount", legend=False, sizes=(20, 2000))

# show the graph
plt.show()
```
