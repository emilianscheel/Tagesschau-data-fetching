```python
import seaborn as sns
import pandas as pd

flights = pd.read_csv('../flights.csv')
flightsPivot = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flightsPivot, linewidths=.5, annot=True, fmt="d", cmap="YlGnBu")
```
